##Average orders

```python
# Group by "Order" and calculate the mean time
grouped = order_df.groupby("Order")["Time"].mean()

# Reset the index to create a DataFrame
order_avg_time = grouped.reset_index()

# Rename the "Time" column to "avg_time"
order_avg_time = order_avg_time.rename(columns={"Time": "avg_time"})

# Select the "Order" and "avg_time" columns
order_avg_time[['Order', 'avg_time']].head()
```

**observation**

*   I calculated the average time for each unique order and presented the results in a data frame with two columns: "Order" and "avg_time." It's useful for understanding the average count of a food item was purchased.



```python
# Calculate the mean time for each "Order" using groupby and mean
grouped = order_df.groupby("Order")["Time"].mean().reset_index()

# Create a larger figure to accommodate the rotated labels
plt.figure(figsize=(12, 6))

# Create the point plot
ax = sns.pointplot(x="Order", y="Time", data=grouped)

# Rotate the y-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

# Add axis labels and a title
plt.xlabel('Order', fontsize=14)  # Increase the font size for the x-axis label
plt.ylabel('Average Time', fontsize=14)
plt.title('Average Time by Order', fontsize=15)

# Adjust the layout to prevent clipping of labels
plt.tight_layout()

# Show the plot
plt.show()


```

Observations -->

* The average times a food got ordered varies a lot were Fried Catfish Hoosier BBQ Pulled Pork Sandwich and Indiana Buffalo Chicken Tacos are around 13.5 times average but Indiana Corn on the Cob and Indiana Pork Chili are below 12 average order times.

# Influence of Year to the order total

```python
# Calculate the mean time for each "Year" using groupby and mean
grouped = order_df.groupby("Year")["Time"].mean().reset_index()

# Create a larger figure to accommodate the rotated labels
plt.figure(figsize=(12, 6))

# Create the point plot
ax = sns.pointplot(x="Year", y="Time", data=grouped)

# Rotate the x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

# Add axis labels and a title
plt.xlabel('Year of Study', fontsize=14)
plt.ylabel('Average Time Ordered', fontsize=14)
plt.title('Influence of Year of Study on Average Time Ordered', fontsize=15)

# Adjust the layout to prevent clipping of labels
plt.tight_layout()

# Show the plot
plt.show()
```

observation >



*   The Freshman year is spending more time on campus with that, they are spending more money on food orders on campus. Therefore, they are the highest average order .

*   There is a significant consistent downfall of average food orders the older students get.That can be correlated to the amount spent on campus.

##Average order by university

```python
# Group the dataset by the 'University' column and calculate the mean of 'Order' for each group
grouped = order_df.groupby("University")["Time"].mean()

# Reset the index of the grouped dataframe so that 'University' becomes a column
university_avg_order = grouped.reset_index()

# Rename the 'Order' column to 'avg_order'
university_avg_order = round(university_avg_order.rename(columns={"Time": "avg_time"}), 2)

# Select only the 'University' and 'avg_order' columns
university_avg_order[['University', 'avg_time']].head()


```

*  This implies that the particular food item was ordered about the indicated numbers of times from each university. Such information may be utilized to determine favourite or more often purchased meals in different universities for example.

```python
import matplotlib.pyplot as plt

# Plot the bar graph
plt.figure(figsize=(8, 6))
plt.bar(university_avg_order['University'], university_avg_order['avg_time'])
plt.xlabel('University', fontsize=14)
plt.ylabel('Average Order Time', fontsize=14)
plt.title('Average Order Time by University', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

**observation**
